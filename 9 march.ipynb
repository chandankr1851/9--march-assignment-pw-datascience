````python
**Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

- **Probability Mass Function (PMF)**: The PMF is a function that describes the probability of a discrete random variable taking on specific values. It assigns a probability to each possible outcome.

Example: The PMF of rolling a fair six-sided die can be represented as follows:
```
PMF(x) = 1/6 for x in {1, 2, 3, 4, 5, 6}
```
This PMF assigns an equal probability of 1/6 to each of the six possible outcomes.

- **Probability Density Function (PDF)**: The PDF is a function that describes the probability distribution of a continuous random variable. It represents the likelihood of the variable taking on values within a range. Unlike the PMF, the PDF does not provide probabilities for specific points but rather for intervals.

Example: The PDF of a standard normal distribution (bell-shaped curve) is represented as a mathematical function and does not assign probabilities to specific values. It describes how the probability is distributed across the entire range of real numbers.

**Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

- **Cumulative Density Function (CDF)**: The CDF is a function that describes the cumulative probability of a random variable being less than or equal to a specific value. It gives the probability that a random variable takes on a value less than or equal to "x."

Example: The CDF of a standard normal distribution can be represented as follows:
```
CDF(x) = Φ(x) = ∫(-∞, x) f(t) dt
```
where Φ(x) is the cumulative probability, f(t) is the PDF of the standard normal distribution, and the integral calculates the area under the PDF curve from negative infinity to "x."

The CDF is used to find probabilities associated with a range of values, calculate percentiles, and make probability statements about random variables.

**Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.**

The normal distribution is commonly used to model situations where data exhibits a bell-shaped or symmetric distribution. Some examples include:

1. **Height of Adults**: Heights of adults in a population often follow a normal distribution. The mean and standard deviation characterize the average height and the spread of heights in the population.

2. **IQ Scores**: IQ scores of a large population tend to be normally distributed, with the mean representing the average intelligence level and the standard deviation indicating the variability in scores.

3. **Measurement Errors**: Errors in scientific measurements, such as those from instruments, often approximate a normal distribution. The mean error represents the bias, and the standard deviation represents measurement precision.

Parameters of the normal distribution:
- **Mean (μ)**: It determines the center or location of the distribution. Shifting the mean shifts the entire distribution along the x-axis.
- **Standard Deviation (σ)**: It controls the spread or variability of the distribution. A larger standard deviation results in a wider distribution.

The normal distribution is completely defined by its mean and standard deviation. Different values of μ and σ will result in different shapes and sizes of normal distributions.

**Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.**

Importance of Normal Distribution:
- It is a fundamental concept in statistics and probability theory.
- Many statistical methods and hypothesis tests are based on the assumption of normality.
- It simplifies data analysis and modeling.

Real-life examples of Normal Distribution:
1. **Grades in a Large Class**: In a large class, student grades often follow a normal distribution, making it easier to set grading curves and assess student performance.

2. **Exam Scores**: Scores on standardized tests, such as the SAT or GRE, are often distributed approximately normally. This helps in comparing individual performance to a reference group.

3. **Height and Weight**: In a population, the distribution of adult heights and weights tends to be normal, making it useful for designing clothing sizes and healthcare guidelines.

4. **Measurement Errors**: Errors in scientific measurements, like those in chemistry or physics experiments, are often normally distributed. Understanding this distribution is essential for assessing measurement accuracy.

**Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?**

- **Bernoulli Distribution**: The Bernoulli distribution models a random experiment with two possible outcomes, typically labeled as "success" and "failure." It is characterized by a single parameter, "p," which represents the probability of success.

Example: Flipping a fair coin, where "heads" (H) is considered success (p = 0.5) and "tails" (T) is considered failure.

- **Binomial Distribution**: The Binomial distribution models the number of successes ("r") in a fixed number of independent Bernoulli trials ("n"), where each trial has the same probability of success ("p"). It is characterized by two parameters, "n" and "p."

Difference:
- Bernoulli Distribution deals with a single trial, while Binomial Distribution deals with multiple trials (repeated Bernoulli trials).
- Bernoulli Distribution is a special case of the Binomial Distribution when "n" is 1.

**Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.**

To find the probability that a randomly selected observation from a normally distributed dataset with a mean (μ) of 50 and a standard deviation (σ) of 10 is greater than 60, you can use the z-score and the standard normal distribution table (or calculator).

1. Calculate the z-score for x = 60 using the formula:
   ```
   z = (x - μ) / σ
   z = (60 - 50) / 10
   z = 1.0
   ```

2. Look up the probability corresponding to the z-score of 1.0 in the standard normal distribution table or use a calculator. The z-score of 1.0 corresponds to a probability of approximately 0.8413.

So, the probability that a randomly selected observation will be greater than 60 is approximately 0.8413 or 84.13%.

**Q7: Explain uniform Distribution with an example.**

- **Uniform Distribution**: The uniform distribution, also known as the rectangular distribution, is a probability distribution where all outcomes are equally likely over a specified range. It is characterized by two parameters: the minimum value (a) and the maximum value (b) within the range.

Example: Rolling a fair six-sided die can be modeled using a discrete uniform distribution. The die has six equally likely outcomes (1, 2, 3, 4, 5, 6), each with a probability of 1/6.

- **Continuous Uniform Distribution**: In the continuous version, all values within a specified interval have equal probability density. For example, selecting a random real number between 0 and 1 from a uniform distribution.

The key characteristic of a uniform distribution is

 that the probability of any specific value within the range is constant and equal.

**Q8: What is the z score? State the importance of the z score.**

- **Z-Score (Standard Score)**: The z-score is a standardized score that measures how many standard deviations a data point is from the mean in a normal distribution. It is calculated using the formula:
   ```
   z = (x - μ) / σ
   ```
   Where:
   - "x" is the data point.
   - "μ" is the mean.
   - "σ" is the standard deviation.

Importance of the z-score:
- It allows data analysts to standardize data from different normal distributions, making comparisons and calculations more meaningful.
- It helps identify how extreme or unusual a data point is within a distribution.
- It is crucial in hypothesis testing, as it enables the calculation of probabilities and confidence intervals.
- Z-scores are used in quality control, finance, and many fields for making data-driven decisions.

**Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

- **Central Limit Theorem (CLT)**: The Central Limit Theorem is a fundamental concept in statistics. It states that, regardless of the underlying population distribution, the distribution of sample means (or sums) from sufficiently large random samples will be approximately normally distributed, as long as certain conditions are met.

Significance of the Central Limit Theorem:
- It allows statisticians to make inferences about population parameters (such as the population mean) based on sample statistics (such as the sample mean).
- It justifies the use of normal distribution-based methods in hypothesis testing and confidence interval estimation.
- It simplifies statistical analysis by providing a common framework for understanding the behavior of sample statistics.
- It is widely applicable in practice, as many real-world phenomena can be approximated by normal distributions through the CLT.

**Q10: State the assumptions of the Central Limit Theorem.**

The Central Limit Theorem (CLT) holds under the following assumptions:

1. **Independence**: The individual observations within each sample must be independent of each other. This assumption ensures that the samples are not influenced by each other.

2. **Random Sampling**: The samples must be drawn randomly from the population of interest. Random sampling ensures that the sample is representative of the population.

3. **Sample Size**: The sample size should be sufficiently large. Although there is no strict rule for what constitutes "sufficiently large," a common guideline is that the sample size should be at least 30. Larger sample sizes tend to provide more reliable results.

4. **Population Distribution**: The underlying population distribution should have a finite mean (μ) and a finite standard deviation (σ). In some cases, the CLT can apply to populations with heavy-tailed distributions, but this may require even larger sample sizes.

5. **No Extreme Skewness**: The population distribution should not be extremely skewed. Extreme skewness can affect the applicability of the CLT.

It's important to note that while the CLT is a powerful tool for approximating the distribution of sample means, it does not guarantee that the resulting distribution will be exactly normal. However, for practical purposes, the approximation is often sufficiently accurate for statistical analysis.
````
